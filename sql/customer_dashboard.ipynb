```python
import pandas as pd

users = pd.read_csv('users (1).csv')
products = pd.read_csv('products (1).csv')

print(users.head())
print(products.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
city_counts = users['city'].value_counts()

print(city_counts)
```

```python
city_counts = users['city'].value_counts()

plt.figure(figsize=(10,5))

city_counts.plot(kind='bar')

plt.title('Users by City')
plt.xlabel('City')
plt.ylabel('Number of Users')

plt.show()
```

```python
city_counts = users['city'].value_counts()

plt.figure(figsize=(10,5))

city_counts.plot(kind='bar')

plt.title('Users by City')
plt.xlabel('City')
plt.ylabel('Number of Users')

plt.savefig('city_analysis.png')

plt.show()
```

```python
age_counts = users['age_group'].value_counts()

plt.figure(figsize=(8,5))

age_counts.plot(kind='bar')

plt.title('Users by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Users')

plt.show()
```

```python
gender_counts = users['gender'].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%'
)

plt.title('Gender Distribution')

plt.show()
```

```python
category_counts = products['category'].value_counts()

plt.figure(figsize=(8,5))

category_counts.plot(kind='bar')

plt.title('Products by Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')

plt.show()
```

```python
brand_rating = products.groupby('brand')['rating'].mean()

plt.figure(figsize=(8,5))

brand_rating.sort_values().plot(kind='bar')

plt.title('Average Rating by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Rating')

plt.show()
```
